### Library

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('K03_counter_strike_2.csv')
data
```

```python
data.describe()
```

```python
data.info()
```

### Mengubah Values Kolom yang bertipe Data String/Objek menjadi Numerik

```python
for col in ['map', 'bomb_planted', 'round_winner']:
    label_encoder = LabelEncoder()
    data[col] = label_encoder.fit_transform(data[col])
```

```python
data
```

### Memisahkan kolom target dari data 

```python
x = data.drop('round_winner', axis = 1)
y = data['round_winner']
```

```python
x
```

```python
y
```

### Splitting Data

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

### Mengetahui Dimensi dari data training 

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

### Modelling KNN

```python
modelKNN = KNeighborsClassifier(n_neighbors=5)
modelKNN.fit(x_train, y_train)
```

```python
hasil_prediksi_KNN = modelKNN.predict(x_test)
```

```python
print('Akurasi Model: ', accuracy_score(y_test, hasil_prediksi_KNN) * 100)
```
